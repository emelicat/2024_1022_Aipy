```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3,3, num=50)

y1 = x * 2 + 1
y2 = x ** 2
figure = plt.figure()
axes = figure.add_subplot()
axes.plot(x,y1,color='red', linestyle='--')
axes.plot(x,y2,color='green', linestyle='-.')
axes.set_xlim(-1,2)
axes.set_ylim(-2,3)
plt.show
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math

t = np.arange(0, 2.5, 0.1)
y1 = np.sin(math.pi * t)
y2 = np.sin(math.pi * t + math.pi/2)
figure = plt.figure()
axes = figure.add_subplot()
axes.plot(t,y1,'b*')
axes.plot(t,y2,'g-')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math

t = np.arange(0, 1, 0.05)
figure = plt.figure()
sin1 = figure.add_subplot(1,2,1)
cos1 = figure.add_subplot(1,2,2)
sin1.plot(t,y1,'b*')
cos1.plot(t,y2,'g-')
plt.show()
```

```python
figure = plt.figure()
axes = figure.add_subplot()
axes.plot([1,2,3,4],[1,4,9,16],'ro')
axes.annotate('first',(1,1.5))
axes.annotate('second',(2,4.5))
axes.annotate('thrid',(3,9.5))
axes.annotate('forth',(4,16.5))
axes.annotate(r'$y = x^2$',(3.1, 12),fontsize=20,bbox={'facecolor':'yellow'})
plt.show
```
